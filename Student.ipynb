```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
grades = [{},{},{}]

for i in range(3):
    grades[i]['Grade'] = np.full(100, i+1)
    grades[i]['Korean'] = np.random.normal(loc=75, scale=5, size=100)
    grades[i]['English'] = np.random.normal(loc=80, scale=10, size=100)
    grades[i]['Math'] = np.random.normal(loc=90, scale=15, size=100)
    
grade_1 = pd.DataFrame(grades[0])
grade_2 = pd.DataFrame(grades[1])
grade_3 = pd.DataFrame(grades[2])    
```

```python
score_a = pd.concat([grade_1, grade_2, grade_3], axis=0, ignore_index=True)
print(score_a)
```

```python
score_a.to_csv('data/student_scores.csv')
```

```python
scores = pd.read_csv("data/student_scores.csv")
```

```python
corr = scores.corr()
print(corr)
```

```python
sns.heatmap(corr, annot=True)
plt.show()
```

```python
print(scores['Korean'].describe())
print(scores['English'].describe())
print(scores['Math'].describe())
```

```python
subject = ['Korean', 'English', 'Math']

for sub in subject:
    plt.hist(score_a[sub], bins = 15)
    plt.title(sub + 'Score Statistics')
    plt.xlabel('Score')
    plt.ylabel('Distribution')
    plt.show()
```
